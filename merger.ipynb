# Merge

```python
# 1. Obtener la lista de nombres de archivos CSV
try:
    # Usa listdir() para obtener todos los elementos y filtra los que terminan en '.csv'
    csv_filenames = [
        f for f in os.listdir(path) if f.endswith('.csv')
    ]
    
    # 2. Construir la ruta completa para cada archivo
    # Es crucial unir el nombre del archivo con la ruta para poder leerlo
    csv_file_paths = [os.path.join(path, f) for f in csv_filenames]

    print(f"Archivos CSV encontrados en '{path}':")
    for file_path in csv_file_paths:
        print(file_path)

except FileNotFoundError:
    print(f"Error: La ruta '{path}' no fue encontrada. Asegúrate de que la carpeta existe.")
```

## Dict

```python
gnr_dict = {
    'Canción': {'idx':1,
                'sub':['Canción','Canción ciudadana','Canción serrana','Canción criolla','Balada',]},
    'Folklore': {'idx':2,
                'sub':['Zamba','Estilo','Tonada','Marcha','Chacarera','Gato',
                       'Huella','Triunfo','Chamamé','Triste campero','Pericón',
                       'Aire de bailecito','Aire de malambo','Aire de zamba',
                       'Canción del litoral','Tonada salteña','Tonada campera',
                       'Chamarrita','Lazo','Bailecito','Rezo gaucho','Sobrepaso',
                       'Cueca','Vidalita',]},
    'Foxtrot': {'idx':3,
                'sub':['Foxtrot','Shimmy','Fox charleston','Fox Bolero',]},
    'Milonga': {'idx':4,
                'sub':['Milonga','Candombe','Milonga candombe','Milongón',
                       'Murga','Murga candombe','Cifra']},
    'Otros ritmos': {'idx':5,
                'sub':['Bolero','Rumba','Guajira','Bambuco','Corrido','Ranchera',
                       'Clásico','Melodía','Shotis','Polca','Fado','Tarantella',
                       'Mazurca','Habanera','Pasodoble','Jota','Java canción']},
    'Poema': {'idx':6,
                'sub':['Poema','Poema lunfardo','Poema evocativo','Recitado cómico',]},
    'Tango': {'idx':7,
                'sub':['Tango','Arr. en tango','Tango balada',
                       'Canción-tango','Tango-estilo','Tango-chamamé']},
    'Vals': {'idx':8,
                'sub': ['Vals','Vals peruano','Arr. en vals','Canción-vals']},
}
```

# Mapping

```python
# Inicializa el diccionario de mapeo Subgénero -> Género Principal
subgenre_to_main_genre = {}

for main_genre, data in gnr_dict.items():
    # Itera sobre la lista de subgéneros ('sub') para cada género principal
    for subgenre in data['sub']:
        # Mapea el subgénero al género principal
        subgenre_to_main_genre[subgenre] = main_genre

print("✅ Diccionario de mapeo inverso creado (primeros 5):")
print({k: subgenre_to_main_genre[k] for k in list(subgenre_to_main_genre)[:5]})
```

```python
import pandas as pd
# Asumiendo que subgnr_df es tu DataFrame

# Usa .map() para asignar el valor del diccionario a la nueva columna
subgnr_df['Género'] = subgnr_df['Subgénero'].map(subgenre_to_main_genre)

print("\n✅ Columna 'Género' creada con éxito.")

# Verifica las primeras filas
print(subgnr_df[['Subgénero', 'Género']].head(10))
```
